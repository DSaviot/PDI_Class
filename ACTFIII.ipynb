```python
#Importar Librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parte I - Preprocesamiento

#Carga de la imagen

image = cv2.imread("ave.jpg")
imagegs = cv2.imread("ave.jpg", cv2.IMREAD_GRAYSCALE)
imagegs_ecualizada = cv2.equalizeHist(imagegs)

hist_original = cv2.calcHist([imagegs], [0], None, [256], [0,256])
hist_ecualizada = cv2.calcHist([imagegs_ecualizada], [0], None, [256], [0,256])

# Mostrar imágenes y histogramas
plt.figure(figsize=(10,5))

# Imagen Original
plt.subplot(2,2,1)
plt.imshow(imagegs, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

# Histograma Original
plt.subplot(2,2,2)
plt.plot(hist_original, color="black")
plt.title("Histograma Original")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")

# Imagen Ecualizada
plt.subplot(2,2,3)
plt.imshow(imagegs_ecualizada, cmap="gray")
plt.title("Imagen Ecualizada")
plt.axis("off")

# Histograma Ecualizado
plt.subplot(2,2,4)
plt.plot(hist_ecualizada, color="black")
plt.title("Histograma Ecualizado")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")

#Imagen Ecualizada en grande
plt.figure(figsize=(10,5))
plt.imshow(imagegs_ecualizada, cmap="gray")
plt.title("Imagen Ecualizada")
plt.axis("off")

plt.tight_layout()
plt.show()
```
